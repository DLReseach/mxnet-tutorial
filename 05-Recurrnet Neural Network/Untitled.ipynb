```python
data = mx.sym.Variable(name='img')
label = mx.sym.Variable(name='softmax_label')
rnn = mx.sym.RNN(data=data, state_size=10, num_layers=1, mode='lstm', name='lstm')
softmaxout = mx.sym.SoftmaxOutput(data=rnn, label=label)
```

```python
mx.viz.plot_network(symbol=softmaxout, shape={'img': (1, 28, 28)})
```
